```python
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
gdf=gpd.read_file('NYC_Taxi_Zones.geojson')
gdf['centroid'] = gdf.geometry.centroid
gdf['centroid_wkt'] = gdf['centroid'].to_wkt()
gdf = gdf.drop(columns=['centroid'])
gdf
```

```python
total_regions = len(gdf)
gdf['population_exchange'] = None
gdf['average_fee'] = None

#以下是一个人口交换的随机表

location_ids = [str(i) for i in range(1, total_regions + 1)]

exchange_matrix = pd.DataFrame(index=location_ids, columns=location_ids, data=np.random.randint(-100, 100, (total_regions, total_regions)))
for i in range(total_regions):
    for j in range(i + 1, total_regions):
        exchange_matrix.iloc[j, i] = -exchange_matrix.iloc[i, j]


average_fee_list = np.random.uniform(10, 20, size=total_regions)

for index, row in gdf.iterrows():
    loc_id = row['location_id']
    if loc_id in location_ids:
        # 由于location_ids是字符串数组，我们可以直接使用它来索引exchange_matrix
        gdf.at[index, 'population_exchange'] = exchange_matrix.loc[loc_id, :].tolist()
        # 对于averageFee，我们需要从location_ids中找到当前loc_id的索引，然后使用这个索引从average_fee_list中获取值
        idx = location_ids.index(loc_id)
        gdf.at[index, 'average_fee'] = average_fee_list[idx]

gdf['population_exchange'] = gdf['population_exchange'].apply(lambda x: str(x))

gdf


```

```python
gdf.to_file('test.geojson',driver='GeoJSON')
```
